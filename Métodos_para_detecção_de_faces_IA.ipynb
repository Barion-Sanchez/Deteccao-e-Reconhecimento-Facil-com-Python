# Detecção de faces com Haar cascade (OpenCV)

```python
imagem = cv2.imread('/content/person.jpg')
```

```python
imagem.shape #Mostra o tamanho da imagem
```

```python
cv2_imshow(imagem) #Mostra a imagem
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY) #Alteração de cor para escala Cinza
```

```python
imagem_cinza.shape #Mostrar quantidade de dados na imagem
```

```python
cv2_imshow(imagem_cinza)
```

```python
detector_facial = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml') #Código para o detector de face
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
```

```python
deteccoes #VIndicam o posicionamento da face na imagem - Eixo X, Y e dimensões da face
```

```python
len(deteccoes) #Mostra quantas faces foram detectadas
```

```python
177 + 197
```

```python
for (x, y, w, h) in deteccoes:
  print(x, y, w, h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 255), 2)
cv2_imshow(imagem)
```
